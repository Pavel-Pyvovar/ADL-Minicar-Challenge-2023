```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
import imgaug.augmenters as iaa
```

```python
augmentations = [
    iaa.Crop(px=(50, 0, 0, 0), keep_size=False),
    iaa.Multiply((0.5, 3.0)),
    iaa.GaussianBlur(sigma=(0.0, 3.0))
]
```

```python
img_augmented = iaa.Sequential(augmentations).augment_image(img)
```

```python
plt.imshow(img_augmented)
```
